```python
import os
import datetime
import random

print("--- Intermediate Python Core Exercise ---")
print("This exercise covers Python fundamentals, data structures, functions,")
print("core libraries (os, datetime, random), file handling, and exception handling.")
print("Estimated completion time: 3-4 hours.")
print("------------------------------------------")

# --- Global Data for the Exercise (Simulated Data) ---
# We'll imagine this is sensor data from various devices.
DEVICE_TYPES = ['Temperature_Sensor', 'Humidity_Sensor', 'Pressure_Sensor', 'Light_Sensor']
LOCATIONS = ['Lab1', 'Lab2', 'OfficeA', 'OfficeB', 'Warehouse']
READING_RANGES = {
    'Temperature_Sensor': (18.0, 30.0), # Celsius
    'Humidity_Sensor': (30.0, 80.0),   # Percentage
    'Pressure_Sensor': (980.0, 1030.0),# hPa
    'Light_Sensor': (100, 1000)        # Lux
}
DEVICE_TYPES = [
    'Temperature_Sensor',
    'Humidity_Sensor',
    'Pressure_Sensor',
    'Light_Sensor'
]

LOCATIONS = [
    'Lab1',
    'Lab2',
    'OfficeA',
    'OfficeB',
    'Warehouse'
]

READING_RANGES = {
    'Temperature_Sensor': (18.0, 30.0),
    'Humidity_Sensor': (30.0, 80.0),
    'Pressure_Sensor': (980.0, 1030.0),
    'Light_Sensor': (100, 1000)
}

raw_sensor_data = []

for i in range(1, 101):
    # Timestamp in the format YYYY-MM-DD HH:MM:SS
    timestamp_str = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # Unique device ID
    device_id = f"DEV{i:03d}"

    # Randomly choose device type
    device_type = random.choice(DEVICE_TYPES)

    # Randomly choose location
    location = random.choice(LOCATIONS)

    # Generate random value based on device type
    min_value, max_value = READING_RANGES[device_type]
    value = round(random.uniform(min_value, max_value), 2)

    # Create a tuple
    reading = (timestamp_str, device_id, device_type, location, value)

    # Add tuple to the list
    raw_sensor_data.append(reading)

# Print all sensor readings
print("\nGenerated Sensor Readings:")
for reading in raw_sensor_data:
    print(reading)
temperature_readings = []

for reading in raw_sensor_data:
    timestamp, device_id, device_type, location, value = reading

    if device_type == "Temperature_Sensor":
        temp_dict = {
            "timestamp": timestamp,
            "device_id": device_id,
            "location": location,
            "temperature_celsius": value
        }

        temperature_readings.append(temp_dict)

# Print the temperature readings
print("\nTemperature Readings:")
for temp in temperature_readings:
    print(temp)
unique_locations = set()

for reading in raw_sensor_data:
    unique_locations.add(reading[3])   # Index 3 is the location

# Print the unique locations
print("\nUnique Locations:")
print(unique_locations)
device_stats = {}

# Initialize the dictionary
for device in DEVICE_TYPES:
    device_stats[device] = (float('inf'), float('-inf'))

# Find the minimum and maximum values for each device type
for reading in raw_sensor_data:
    device_type = reading[2]
    value = reading[4]

    min_value, max_value = device_stats[device_type]

    if value < min_value:
        min_value = value

    if value > max_value:
        max_value = value

    device_stats[device_type] = (min_value, max_value)

# Display the results
print("\nDevice Statistics:")
for device, stats in device_stats.items():
    print(device, ":", stats)
print("\n--- Task 1.5: Advanced Indexing and Slicing (Lists and Strings) ---")
sample_string = "PythonProgrammingIsFun"
# Extract elements from the 10th to the 20th (inclusive)
sliced_readings_1 = raw_sensor_data[9:20]

# Extract every 5th reading starting from the beginning
sliced_readings_2 = raw_sensor_data[::5]

# Print the results
print("\nReadings from 10th to 20th (inclusive):")
for reading in sliced_readings_1:
    print(reading)

print("\nEvery 5th reading:")
for reading in sliced_readings_2:
    print(reading)
# Extract the substring "Programming"
substring_1 = sample_string[6:17]

# Extract the reverse of "Python" -> "nohtyP"
substring_2 = sample_string[5::-1]

# Print the results
print("\nSubstring 1:", substring_1)
print("Substring 2:", substring_2)
# Access the device_type of the 5th reading
fifth_reading_device_type = raw_sensor_data[4][2]

# Extract the last 6 characters
device_type_suffix = fifth_reading_device_type[-6:]

# Print the results
print("\n5th Reading Device Type:", fifth_reading_device_type)
print("Device Type Suffix:", device_type_suffix)
# 2.1 Function to get readings by location
print("\n--- Task 2: Function Definitions and Usage ---")

def get_readings_by_location(readings_list, target_location):
    result = []

    for reading in readings_list:
        if reading[3] == target_location:
            result.append(reading)

    return result


# Example usage
location = "Lab1"
lab1_readings = get_readings_by_location(raw_sensor_data, location)

print(f"\nSensor Readings for {location}:")
for reading in lab1_readings:
    print(reading)
# 2.2 Function to calculate average temperature

def calculate_average_temperature(temp_data_list):
    if len(temp_data_list) == 0:
        return 0.0

    total = 0

    for reading in temp_data_list:
        total += reading["temperature_celsius"]

    average = total / len(temp_data_list)
    return round(average, 2)


# Example usage
avg_temperature = calculate_average_temperature(temperature_readings)

print("\nAverage Temperature:", avg_temperature)
# 2.3 Function to generate a formatted report line

def generate_report_line(reading_dict):
    return f"[{reading_dict['timestamp']}] Device {reading_dict['device_id']} at {reading_dict['location']}: {reading_dict['temperature_celsius']}°C"


# Example usage
if len(temperature_readings) > 0:
    report = generate_report_line(temperature_readings[0])
    print("\nReport Line:")
    print(report)
```
