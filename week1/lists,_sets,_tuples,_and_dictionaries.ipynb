```python
fruits=['apple','orange','pineapple','cherry']
print("orginal list:",fruits)
print("second fruit",fruits[1])
fruits.append('grapes')
print("after appending:",fruits)
fruits.remove('orange')
print("after removing:",fruits)
print("no of fruits:",len(fruits))
numbers=(1,2,9)
print("orginal:",numbers)
print("second element:",numbers[1])
unique_numbers={1,2,3,2,4,4}
print("set(no duplicates):",unique_numbers)
unique_numbers.add(7)
print("after adding:",unique_numbers)
unique_numbers.remove(4)
print("after removing:",unique_numbers)
set_a={1,2,3}
set_b={2,3,4,5}
print("union:",set_a | set_b)
print("intersection:",set_a & set_b)
print("difference:",set_a-set_b)
student={
    'name':'john',
    'age':21,
    'course':'python'
}
print("dictionary:",student)
print("student name:",student['name'])
student['grade']='O'
print("after grade",student)
del student['age']
print("after deleting",student)
for key,value in student.items():
  print(f"{key}: {value}")
  keys=['id','name','marks']
  values = [101, 'John', 88]
student_info = dict(zip(keys, values))
print("Student Info from zip:", student_info)
```
